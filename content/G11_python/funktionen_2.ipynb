## Rückgabewerte

Manchmal möchte man mehr als einen Wert zurückgeben. Das könnte z.B. nötig sein,  
wenn deine Funktion den größten und kleinsten Wert einer Liste ermitteln soll.  
Je nach Programmiersprache ist das etwas anders.  
In Python kannst du mehrere Werte zurückgeben und bekommst etwas, das einer Liste  
sehr ähnlich ist. Betrachte und analysiere die nachfolgenden Beispiele:

```python
def randwerte(zahlenliste):
    vorne = zahlenliste[0]
    hinten = zahlenliste[-1]
    return vorne, hinten

# Ausführung
version_1 = randwerte([7,8,9])
v = version_1[0]
h = version_1[1]
#oder
v, h = randwerte([7,8,9])
```
