## Was ist eine Liste?

Eine Liste ist wie eine geordnete Sammlung von Dingen. In Python steht sie in eckigen Klammern und die Elemente sind durch Kommas getrennt.

### Beispiel
```python

woerter = ["rot", "grün", "blau"]
gemischt = ["Anna", 17, True]  # möglich, aber am Anfang lieber gleiche Typen verwenden
```

## Auf einzelne Elemente zugreifen

Die Positionen in der Liste (= Index) beginnt bei 0!

```python
farben = ["rot", "grün", "blau"]
print(farben[0])  # erstes Element: "rot"
print(farben[1])  # zweites Element: "grün"
print(farben[-1]) # letztes Element: "blau"
```

**Achtung:**  
`IndexError` tritt auf, wenn der Index zu groß/klein ist.

## Elemente ändern

Du kannst ein Element per Index ersetzen:

```python

print(zahlen)  # [5, 60, 7]
```

## Elemente hinzufügen

    append(x) hängt x hinten an.
    insert(pos, x) fügt x an Position pos ein.

```python
obst = ["Apfel", "Banane"]
obst.append("Kirsche")
print(obst)  # ["Apfel", "Banane", "Kirsche"]

obst.insert(1, "Birne")
print(obst)  # ["Apfel", "Birne", "Banane", "Kirsche"]
```

## Elemente entfernen

    pop() entfernt das letzte Element und liefert es zurück.
    pop(i) entfernt Element an Position i und liefert es zurück.
    remove(x) entfernt das erste Vorkommen von x (Fehler, wenn x nicht da ist).

```python
farben = ["rot", "grün", "blau", "grün"]
letztes = farben.pop()
print(letztes)   # "grün"
print(farben)    # ["rot", "grün", "blau"]

farben.remove("grün")
print(farben)    # ["rot", "blau"]
```

## Länge bestimmen

Mit `len(liste)` bekommst du die Anzahl der Elemente.

```python
tiere = ["Hund", "Katze", "Maus"]
print(len(tiere))  # 3
```

## Über Listen laufen (Iteration)

Mit einer for-Schleife besuchst du jedes Element nacheinander:

```python
for f in ["rot", "grün", "blau"]:
    print("Farbe:", f)
```

Manchmal brauchst du auch die Position. Dann nutze `range` oder `enumerate`:

```python
tiere = ["Hund", "Katze", "Maus"]
for i in range(len(tiere)):
    print(i, tiere[i])

for i, tier in enumerate(tiere):
    print(i, tier)
```

## Teilstücke (Slicing)

Mit `liste[start:stop]` holst du einen Ausschnitt (bis vor stop):

```python
zahlen = [10, 20, 30, 40, 50]
print(zahlen[1:3])  # [20, 30]
print(zahlen[:2])   # [10, 20] (von Anfang)
print(zahlen[2:])   # [30, 40, 50] (bis Ende)
```

## Weitere nützliche Operationen

    x in liste: prüft, ob x enthalten ist.
    liste.count(x): wie oft kommt x vor?
    liste.index(x): an welcher Position steht x? (Fehler, wenn nicht da)
    liste.sort(): sortiert die Liste (nur sinnvoll, wenn alle Elemente vergleichbar sind, z. B. nur Zahlen oder nur Strings).

```python
namen = ["Anna", "Ben", "Anna"]
print("Anna" in namen)   # True
print(namen.count("Anna"))  # 2
print(namen.index("Ben"))   # 1
namen.sort()
print(namen)                # ['Anna', 'Anna', 'Ben']
```
